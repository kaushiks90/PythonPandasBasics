```python
import pandas as pd
```

```python
df1=pd.read_csv("train.csv")
df1.head()
```

```python
gender={"male":1,"female":0}
df1["gender"]=df1.Sex.map(gender)
df1.head()
```

```python
df1.Pclass=df1.Pclass.map(lambda x:x+3)
df1.head()
```

```python
capitalize=lambda x:x.upper()
df1.Name=df1.Name.apply(capitalize)
df1.head()
```
